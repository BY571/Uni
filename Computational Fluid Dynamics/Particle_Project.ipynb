```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
import time

%matplotlib qt4
```

```python
class Particle():
    def __init__(self,u,v,xp,name):
        
        self.name = name
        self.u = u
        self.v = v
        self.xp = xp
        self.x = np.random.rand() #uniform distribution over [0, 1)
        self.y = np.random.rand() #uniform distribution over [0, 1)
        self.new_xp = 0
        
    def return_particle_name(self):
        return str(self.name)
    
    def give_x_y(self):
        return self.x, self.y
        
    def return_all(self):
        return self.u, self.v, self.xp
    
    def calc_uv(self):
        self.u = math.sin(4*math.pi*self.y) # switched x,y
        self.v = math.sin(4*math.pi* self.x - math.pi/2)+0.4
        
    def give_uv(self):
        return self.u, self.v
    
    def give_xp(self):
        return self.xp
    
    def calc_new_xp(self, dt):
        self.new_xp = self.xp + dt*np.array([[self.u],[self.v]])
        
    def return_new_xp(self):
        return self.new_xp
    
    def swap_xp(self):
        self.xp = self.new_xp
        self.x = self.xp[0][0]
        self.y = self.xp[1][0]
```

```python

x = np.array([[0],[0]]) # Anfangsbedingung
u_start = 0
v_start = 0
p = Particle(u_start,v_start,x)
```

```python
# Testing:
print("Start_Position: ",p.give_x_y())
print("Old_xp: \n", p.give_xp())
p.calc_uv()
p.calc_new_xp(0.5)
print("-------------------------------")
print(p.return_new_xp())
print("-------------------------------")
p.swap_xp()
print("New_Position: ", p.give_x_y())

print("New_xp: \n",p.give_xp())
```
