```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_excel("Vorort Messungen.xlsx")
data.head()
```

```python
y = data["Tiefe"]
temperatur = data["T in C°"]
el = data['EL in MS/cm']
ph = data["Ph"]
o2 = data["O² mg/l"]

```

```python
plt.plot(y, temperatur, "r", y, el, "g", y,ph,"b", y,o2,"y")
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,10))

plt.subplot(221)
plt.plot(y, temperatur, 'r')
plt.xlabel('T in [C°]')
plt.ylabel('Tiefe in [m]')
plt.grid(True)
plt.subplot(222)
plt.plot(y, el, 'y')
plt.xlabel('EL in [MS/cm]')
plt.ylabel('Tiefe in [m]')
plt.grid(True)

plt.subplot(223)
plt.plot(y, ph, 'b')
plt.xlabel('Ph')
plt.ylabel('Tiefe in [m]')
plt.grid(True)
plt.subplot(224)
plt.plot(y, o2, 'g')
plt.xlabel('O² [mg/l]')
plt.ylabel('Tiefe in [m]')

plt.grid(True)
plt.subplots
#plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    #wspace=0.35)

plt.show()
```

```python
plt.rcParams['xtick.top'] = plt.rcParams['xtick.labeltop'] = True
plt.rcParams['xtick.bottom'] = plt.rcParams['xtick.labelbottom'] = False

fig, ax = plt.subplots(figsize = (8,8))
plt.subplot(111)
plt.plot(temperatur, y, 'r', o2,y,"b", ph,y,"g")
plt.ylim(8,0) # decrease

plt.suptitle('Temperatur in [C°]')
plt.ylabel('Tiefe in [m]')
plt.legend(["Temperatur","O2 [mg/l]","Ph-Wert"],loc = 10)

plt.grid(True)

plt.show()
```
