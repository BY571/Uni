
### Erklärung des Verlaufs der n-M-Kennlinie

Nach der Formel für das innere Moment steigt das Moment proportional zum Ankerstrom. <br/>
$M_{i} = c \Phi \cdot \ i_{a}$ <br/>
Bei starken Belastungen, also hohen Ankerströmen, wird ein starkes Ankerquerfeld hervorgerufen. Anhand dessen kann es zu nicht gewollten Drehzahlerhöhungen kommen. Der Effekt der Drehzahlerhöhung ist also eine Folge der Ankerrückwirkung.

### Erklärung des Verlaufs der $ \ I_{A}-M$ - Kennlinie

Der Verlauf der $\ I_{A}-M-Kennlinie$ ist auf die Formel des inneren Moments zurückzuführen. <p>
$M_{i} = c \Phi \cdot \ i_{a}$ <p>
Demnach steigt das Moment proportional zum Ankerstrom.

### Erklärung des Verlaufs der $\eta-P$ - Kennlinie bei Ankervorwiderstand

Anhand der (Geraden-) Gleichung zur Drehzahlstellung ist die Abhängigkeit zwischen Drehzahl und Drehmoment zu erkennen <p>
$\omega = \frac{U}{c\Phi}- \frac{R_{a}+R_{v}}{(c\Phi)^2} M_{i}$ <p>
Wird nun ein zusätzlicher Ankervorwiderstand geschaltet ($\ R_{v}$) dann erhöht sich der Betrag der Steigung in der Geradengleichung für die Drehzahl. Das hat zur Folge, dass die Drehzahl mit Ankervorwiderstand bei steigendem Drehmoment stärker abfällt.<br/>
Mit der Formel $\ P = M \cdot \omega = M \cdot 2\pi n$ für das zu einer geringeren Leistung und schlussendlich durch die Verluste zu einem geringeren Wirkungsgrad, da die Verluste im Ankervorwiderstand nach $P_{v} = I_{A}^2 \cdot R_{V}$ in Wärme umgewandelt werden.

### Erklaerung zum Verlauf der $I_{A}-M$ - und $ n-M$ - Kennlinie

Das innere Moment der Reihenschlussmaschine ist definiert über: <p>
$ M = c_{r}\cdot i^2 $ <p>
nach dem Umstellen nach I ergibt dies: <p>
$i= \sqrt{\frac{M}{c_{r}}} $ <p>
demnach besitzt die Kennlinie einen parabolischen Verlauf.
<p>
Der Verlauf der Drehzahl-Drehmoment-Kennlinie ergibt sich aus der Formel:<p>
$U_{q} = c_{r} i \omega$ <p>
stellt man diese Gleichung nach dem Strom i um und setzt sie in die obere Gleichung für das Drehmoment ein erhält man die Formel für den Verlauf der Drehzahl-Drehmoment-Kennlinie <p>
$M =\frac{U^2_{q}}{c_{r}} \cdot \frac{1}{\omega^2}$ <p> bzw.   $\omega = \sqrt{\frac{U^2_{q}}{c_{r}}\cdot \frac{1}{M}}$ <p>
welche eine starke Lastabhängigkeit und einen hyperbolischen Drehzahlverlauf aufzeigt.

### Erklaerung des Einsatzes der GRM für Triebfahrzeuge und Aufzuege

Anhand des Drehzahl-Drehmoment-Verlaufs ist zu erkennen, dass bei niedriger Drehzahl maximales Moment bzw. bei niedriger Belastung unzulässig hohe Drehzahl anliegt. Daher werden GRM für Triebfahrzeuge und Aufzüge benutzt, da hier immer eine Belastung auftritt und die Drehzahl nie in den kritischen Bereich kommt.
